```python
import numpy as np
import pandas as pd
```

```python
# zajem strani

url = 'https://www.bsi.si/en/statistics/exchange-rates-and-precious-metals-prices/bank-of-slovenia-exchange-rates-ecb-reference-rates'
df = pd.read_html(url)[0]
df
```

```python
# zabeležimo vrstni red in vrednosti za eur v array

vrstni_red = df['Currency'].to_numpy()
eur = df['Rate'].to_numpy()

vrstni_red = np.append(vrstni_red, 'EUR')
eur = np.append(eur, 1.00)
eur
```

```python
# naredimo matriko za Floyd-Warshalla

n = len(eur)
rates = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        rates[i, j] = eur[j] / eur[i]

# krat v plus in max v min
rates = - np.log(rates)
rates
```

```python
def floyd_warshall(utezi):
    """
    Izvede Floyd-Warshallov algoritem za graf z danimi utežmi.

    utezi = matrika, ki predstavlja graf
    """
    # naredimo matriko za začetek -> inf matrika
    n = len(utezi)
    dist = np.zeros((n, n))
    dist.fill(np.inf)

    # to je nepotrebno -> samo začni z utezi
    # mogoče bolj splošno
    for i in range(n):
        for j in range(n):
            if i == j:
                dist[i, j] = 0
            elif utezi[i, j] < np.inf:
                dist[i, j] = utezi[i, j]

    # dejanski algo      
    for k in range(n):
        for i in range(n):
            for j in range(n):
                if dist[i, j] > dist[i, k] + dist[k, j]:
                    dist[i, j] = dist[i, k] + dist[k, j]
    
    return dist
```

```python
optimum = floyd_warshall(rates)
np.diag(optimum)
```

```python
tes = np.array([[0, np.inf, -2, np.inf], [4, 0, 3, np.inf], [np.inf, np.inf, 0, 2], [np.inf, -1, np.inf, 0]])
print(floyd_warshall(tes))
```
